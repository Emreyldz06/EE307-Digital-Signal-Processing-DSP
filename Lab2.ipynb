Create and plot the following signal when $f_1=10$ Hz, take t between 0 and 0.5. (Determine the step size such that it resembles continous time signal) 

$ x_1(t)=cos(2 \pi f_1 t)$

```python
# Write your code here
import matplotlib.pyplot as plt
import numpy as np
f=10
fs=300
t=np.arange(0,0.5+1/fs,1/fs)
x1=np.cos(2*np.pi*f*t)
plt.figure()
plt.title("Example Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x1)
plt.show();
```

We will use the samples that we have obtained in the previous part to reconstruct our original signal. In Python, we can visualize the reconstructed signal by using plt.plot(). It is the interpolated signal using the samples. So plot the reconstructed signal. Does it resemble the original signal?
```python
# Write your code here
import matplotlib.pyplot as plt
import numpy as np
f=10
fs=12
t=np.arange(0,0.5833,1/fs)
x1=np.cos(2*np.pi*f*t)
plt.figure()
plt.title("Reconstructed Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x1,".-")
plt.show()
```

I used the plt.plot method to reconstruct sampled signal. The reason we couldn't observed a 10 Hz sinusoidal signal is our sampling frequency is lower than 2.f (20Hz). In this situation the aliasing occured and we lost the information about the original signal.

## PART B
In this part, we will see the effects of aliasing and undersampling.

Consider the two sinusoidal signals:
$ x_1(t)=cos(2 \pi f_1 t)$ and $ x_2(t)=cos(2 \pi (f_1+f_2) t)$ where $f_2>2 f_1$. We will show that we cannot differentiate these signals if we sample them using the sampling frequency $f_s=f_2$. 

First plot the two signals when $f_1=10$ Hz and $f_2=100$ Hz in the same time interval (e.g between 0 and 0.2) on the same figure. You should plot signals in different colors and add legend to the figure. 

```python
# Write your code here
import matplotlib.pyplot as plt
import numpy as np
f1=10
f2=100
fs=10000 #To show two signas are different.  
t=np.arange(0,0.2+1/fs,1/fs)

x1=np.cos(2*np.pi*f1*t)
x2=np.cos(2*np.pi*(f1+f2)*t)

plt.plot(t,x1,'r',label='10 Hz')
plt.plot(t,x2,label='110 Hz')
plt.legend(loc='lower right')
plt.show()
```

Firstly,I set the sampling value is enough high to show that two signals are obviously different from each other. As a result, I showed theese two signals with legend command via matplotlib.pyplot library. 

As you have done in the first part, take samples from two signals, visualize these discrete-time signals $x_1(nT_s)$ and $x_2(nT_s)$ with plt.stem() function and try to reconstruct the original signal so plot interpolated signals with plt.plot().

```python
# Write your code here
import matplotlib.pyplot as plt
import numpy as np
f1=10
f2=100
fs=100
t=np.arange(0,0.2+1/fs,1/fs)

x1=np.cos(2*np.pi*f1*t)
x2=np.cos(2*np.pi*(f1+f2)*t)

plt.figure()
plt.title("Reconstracted 10Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x1,"r")
plt.figure()
plt.title("Sampled 10 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.stem(t,x1,"r")
plt.show()
plt.figure()
plt.title("Reconstracted 110 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x2,)
plt.figure()
plt.title("Sampled 110 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.stem(t,x2)
plt.show()
```

The reason we see same signals when we reconstruct theese two signals is sampling frequency is not satisfies the Nyquist Theroem. Since sapmling rate is not satisfies the Nyquist Rate (fs>=2fm) the aliasing is occurs. Thats why when signals are reconstructed we observe the same spectrums. Also the major cause of this situation is fs=f2.

## Part B

Repeat part (a) for the triangular window:
$$w_\triangle [n] =  \left ( 1- \frac{|n|}{N} \right ) w[n].$$



```python
# write your code here 
# Basically you will change the window function in the first part.
i=1
plt.figure(figsize=(15,15))
for N in [5, 20, 100]:  

    
        # create 2N+1-point temporal indices (NN) from -N to N using np.arange
        # write your code here
        NN_t = np.arange(-N,N+1,1)
        # create window function (w) using np.ones
        # write your code here
        triList = [1-abs(n)/N for n in range(-N, N+1)]
        w_t = np.array(triList)
        # create 512-point frequency indices (Omega) from -pi to pi (use np.pi)
        # write your code here
        Omega_t = np.linspace(-np.pi, np.pi, 512)
        # create matrix of appropriate temporal indices (n) and frequency indices (omega) using np.meshgrid
        # write your code here
        n_t ,omega_t= np.meshgrid(NN_t, Omega_t)
        # create DTFT matrix 
        DTFT = np.exp(-1j*n_t*omega_t)
        
        # evaluate DTFT
        W = DTFT.dot(w_t)
        # normalize W such that W at omega=0 is 1
        # write your code here
        normalize = 1/W[int((len(Omega)-1)/2)]
        W = W*normalize
        # create plots
        # plot both real and imag on the same figure ( use W.real and W.imag )
        # write your code here
        plt.subplot(2,3,i)
        plt.title('DTFT Real',size=15)
        plt.xlabel('Omega',size=15)
        plt.ylabel('Amplitude',size=15)
        plt.plot(Omega, W.real,"r")
        plt.subplot(2,3,3+i)
        plt.title('DTFT imag',size=15)
        plt.xlabel('omega',size=15)
        plt.ylabel('Amplitude',size=15)
        plt.plot(Omega, W.imag,"k")
        
        
        # label plots and put title and legends ( use plt.xlabel, plt.title, plt.legend )
        # write your code here
        i+=1
      
```

## Part C

Consider the ideal low-pass filter with cut-off frequency $\omega_c = 0.3 \pi$. Suppose that we will truncate the impulse response of this filter using the windows in parts (a) and (b). Hint: Windowing a signal in time domain means that we multiply the signal with the windowing function. Since you have frequency domain representation of rectangular and triangular windows, you can do the operation in frequency domain.

Plot the frequency response of the resulting filters on the same figure. Don't forget to add legend otherwise it cannot be understood clearly. Explain the differences. Which one is advantageous and why? Justify your reasoning.



```python
# write your code here 
i=1
plt.figure(figsize=(15,15))
for N in [5, 20, 100]:  
    
        
        NN = np.arange(-N,N+1,1)
        
        w = np.ones(2*N + 1)

        h1 = [np.sin(0.3 * np.pi * n)/(np.pi * n) for n in range(-N, 0)]
        h1.append(1)
        h2 = [np.sin(0.3 * np.pi * n)/(np.pi * n) for n in range(1, N+1)]
        hList = h1+h2
        h = np.array(hList)*w
    
        Omega = np.linspace(-np.pi, np.pi, 512)
        n ,omega= np.meshgrid(NN, Omega)
  
        DTFT= np.exp(-1j*n*omega)

        W = DTFT.dot(h.T)
        normalize = 1/W[int((len(Omega)-1)/2)]
        W = W*normalize
        
        plt.subplot(2,3,i)
        plt.title('DTFT Real',size=15)
        plt.xlabel('Omega',size=15)
        plt.ylabel('Amplitude',size=15)
        plt.plot(Omega, W.real,"r")
        plt.subplot(2,3,3+i)
        plt.title('DTFT imag',size=15)
        plt.xlabel('omega',size=15)
        plt.ylabel('Amplitude',size=15)
        plt.plot(Omega, W.imag,"k")

        i+=1
```

The FIR filter effect is observed. The energy of the signal has been reduced with the low pass filter and the correction has been made.As the sample number increases, a more stable signal is obtained, so the C part with a high N number is the most advantageous.

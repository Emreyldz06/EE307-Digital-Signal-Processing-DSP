```python
def mysinewave(freq,step_size,max_duration,amp):
    
    t=np.arange(0,max_duration,step_size)
    signal=amp*np.sin(2*np.pi*freq*t)
    
    return t,signal
```

In the part given above, The mysinewave function is defined to write sine function just one time.
This function takes parameters freq=frequency,step_size= step size and max_duration= Ending point for to np.arange function
finally asks for paramter as amp to describe amplitude.
It returns time and the signal.

```python
t1,signal1=mysinewave(16,0.05,1,6);
t2,signal2=mysinewave(16,0.01,0.99,6);
t3,signal3=mysinewave(16,0.001,0.999,6);
```

In the part written above I called mysinewave function 3 times. 
This function takes parameters according to the parmeters given to me in Assingment1 document. (According to my student number)


```python
plt.figure()
plt.title("Signal 1")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.plot(t1,signal1,'g')
plt.show()
```

Since the sampling frequency is 20Hz( Ts=0.05s) which is lower than Nyquist rate, our output is similar to sine wave but it didn't converge to smooth sine wave.The reason for its looks like a sharp sine wave is plt.plot() function complates the each discreate sampled points.   

```python
plt.figure()
plt.title("Signal 2")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.plot(t2,signal2,'r')
plt.show()
```

In this part it seems like that frequency is increased. However, frequency is fixed. The reason for this is our sampling rate is increased by decrasing step size of np.arange function. Therefore, we observed more sampled points and it looks like sharp sine wave. The reason for sharpness is as I stated before plt.plot() function draws line between two sampled point directly. 

```python
plt.figure()
plt.title("Signal 3")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.plot(t3,signal3,'b')
plt.show()
```

In the last part, we increased the sampling rate enough to see a smooth sinusoidal signal.
